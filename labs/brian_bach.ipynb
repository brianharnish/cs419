```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import tree
```

```
chords = pd.read_csv('https://raw.githubusercontent.com/zacharski/ml-class/master/data/bach.zip')
chords.set_index('choral_ID', inplace=True)
chords = chords.drop('event_number', axis=1)
chords
```

```
one_hot = pd.get_dummies(chords['bass'], prefix=["bass"])
chords = chords.drop('bass', axis=1)
chords = chords.join(one_hot)
chords
```

```
array = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
for i in array:
  chords[i] = chords[i].apply(lambda x: 1 if x == 'YES' else 0)
chords
```

```
cFeatures = chords.drop('chord_label', axis=1)
cLabels = chords['chord_label']
chord_train_features, chord_test_features, chord_train_labels, chord_test_labels = train_test_split(cFeatures, cLabels, test_size = 0.2, random_state=42)
clf= tree.DecisionTreeClassifier(criterion= 'entropy')
clf.fit(chord_train_features, chord_train_labels)
cross_scores = cross_val_score(clf, chord_train_features, chord_train_labels, cv=10)

print(cross_scores)
print('Mean is {}'.format(cross_scores.mean()))
```

```
from sklearn.metrics import accuracy_score
predictions = clf.predict(chord_test_features)
print(predictions)
accuracy_score = accuracy_score(chord_test_labels, predictions)
print(accuracy_score)
compare = pd.DataFrame(chord_test_labels)
compare['predicted'] = predictions
compare
```
